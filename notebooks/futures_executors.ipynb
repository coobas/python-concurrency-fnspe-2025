```python
from concurrent.futures import ProcessPoolExecutor, ThreadPoolExecutor, InterpreterPoolExecutor

MAX_WORKERS = 4
```

```python
process_executor = ProcessPoolExecutor(max_workers=MAX_WORKERS)
thread_executor = ThreadPoolExecutor(max_workers=MAX_WORKERS)
interpreter_executor = InterpreterPoolExecutor(max_workers=MAX_WORKERS)
```

```python
import random
from functools import partial
```

```python
list(process_executor.map(partial(random.randint, 0), 8*[100]))
```

```python
thread_executor = ThreadPoolExecutor(max_workers=MAX_WORKERS)
list(thread_executor.map(partial(random.randint, 0), 8*[100]))
```

```python
interpreter_executor = InterpreterPoolExecutor(max_workers=MAX_WORKERS)
list(interpreter_executor.map(partial(random.randint, 0), 8*[100]))
```

```python
import my_random

print(list(process_executor.map(my_random.safe_randint, 8*[100])))
print(list(thread_executor.map(my_random.safe_randint, 8*[100])))
print(list(interpreter_executor.map(my_random.safe_randint, 8*[100])))
```

```python
import numpy as np

interpreter_executor.submit(np.random.randint, 10).result()
```
